```python
% Zadanie 1

% Fakty
ojciec(tomek, jan).
ojciec(tomek, olek).
ojciec(jan, anna).
ojciec(jan, piotr).
ojciec(olek, adam).
matka(halina, adam).
matka(maria, anna).
matka(maria, piotr).
malzenstwo(jan, maria).
mezczyzna(jan).
mezczyzna(piotr).
kobieta(maria).
kobieta(anna).

% Reguła rodziców
rodzic(X, Y) :- ojciec(X, Y).
rodzic(X, Y)  :- matka(X, Y).

% Reguła dla relacji przodek
przodek(X, Y) :- rodzic(X, Y).
przodek(X, Y) :-
    rodzic(X, Z),
    przodek(Z, Y).
%Pytania i odpowiedzi: ?- przodek(tomek, adam). - true; ?- przodek(adam, anna). - false

%Reguły dla relacji siostra, brat
brat(X, Y):-
    rodzic(Z, X),
    rodzic(Z, Y),
    X \= Y,
    mezczyzna(X).

siostra(X, Y) :-
    rodzic(Z, X),
    rodzic(Z, Y),
    X \= Y,
    kobieta(X).
%Pytania i odpowiedzi: ?- brat(piotr, anna). - true; ?- brat(piotr, piotr). - false

%Reguła dla relacji kuzyn
kuzyn(X, Y) :-
    rodzic(Z, X),
    rodzic(W, Y),
    (brat(Z, W); siostra(Z, W)),

kuzyn(X, Y) :- kuzyn(Y, X).

%Pytania i odpowiedzi: ?- kuzyn(X, piotr). - X = adam
```

```python
% Zadanie 2

% Fakty
lotnisko(warszawa).
lotnisko(krakow).
lotnisko(gdansk).
lotnisko(wroclaw).
lotnisko(poznan).
polaczenie(warszawa, krakow, 300).
polaczenie(warszawa, gdansk, 500).
polaczenie(warszawa, wroclaw, 150).
polaczenie(wroclaw, gdansk, 100).
polaczenie(krakow, poznan, 150).
polaczenie(poznan, gdansk, 250).

% Reguła na bezpośrednie połączenie lotnicze między lotniskiem X a lotniskiem Y.
polaczenie(X, Y) :- polaczenie(X, Y, _).
% Pytania i odpowiedzi: ?- polaczenie(warszawa, krakow). - true; ?- polaczenie(gdansk, krakow). - false

% Reguła określa koszt podróży między lotniskiem X a lotniskiem Y.
koszt_podrozy(X, Y, C) :- polaczenie(X, Y, C).
% Pytania i odpowiedzi: ?- koszt_podrozy(warszawa, gdansk, C). - 500;

% Najkrótsza droga: Jeżeli istnieje bezpośrednie połączenie
najkrotsza_droga(X, Y, C) :- polaczenie(X, Y, C).

% Najkrótsza droga: Rekurencyjnie sprawdzamy wszystkie połączenia pośrednie
najkrotsza_droga(X, Y, C) :-
    polaczenie(X, Z, C1),
    najkrotsza_droga(Z, Y, C2),
    C is C1 + C2.

% % Pytanie i odpowiedź: ?- najkrotsza_droga(krakow, gdansk, C). - 400
```

```python
%Zadanie 3

% Fakty
produkt(tv, 10, sektor_a).
produkt(pralka, 5, sektor_b).
produkt(lodowka, 3, sektor_c).
:- dynamic produkt/3.     % Mówię, że produkt/3 jest dynamiczny.

% Regułą sprawdza, czy produkt o nazwie X jest dostępny w magazynie
dostepny_produkt(X) :-
    produkt(X, Ilosc, _),
    Ilosc > 0.
%Pytanie i odpowiedź: ?- produkt(tv, 10, sektor_a). - true

% Reguła przenosi produkt o nazwie X do innego sektora Y magazynu.
przenies_produkt(X, Y) :-
    produkt(X, Ilosc, _),           % Pobierz produkt X i jego ilość
    asserta(produkt(X, Ilosc, Y)),  % Dodanie nowego faktu z nową lokalizacją
    retract(produkt(X, Ilosc, _)),  % Usunięcie starego faktu
    write(product(X, Ilosc, Y)).
% Pytanie i odpowiedź: ?- product(tv,10,sektor_b) true

% Reguła do uzupełniania stanu produktu
uzupelnij_stan(X, N) :-
    produkt(X, Y, Z),             % Pobierz produkt X i jego ilość
    W is Y + N,                   % Dodaj ilości
    retract(produkt(X, Y, Z)),    % Usunięcie starego faktu
    asserta(produkt(X, W, Z)),    % Dodanie nowego faktu z nową ilością
    write(produkt(X, W, Z)).
% Pytanie i odpowiedź: ?- uzupelnij_stan(pralka, 3). - produkt(pralka,8,sektor_b) true
```

```python
% Zadanie 4

%Reguły
miasto(warszawa).
miasto(krakow).
miasto(poznan).
droga(warszawa, krakow, 4).
droga(krakow, poznan, 3).
droga(poznan, warszawa, 7).

droga_bezposrednia(X, Y) :- droga(X, Y, _). % Sprawdzenie, czy istnieje bezpośrednie połączenie
% Reguła dla drogi X -> Y
droga_bezposrednia(X, Y, T) :-
    droga(X, Y, T).
% Pytanie i odpowiedź: ?- droga_bezposrednia(warszawa, krakow). - true

% Regułę dla drogi Y -> X, na kierunek odwrotny
droga_bezposrednia(X, Y, T) :-
    droga(Y, X, T).
% Reguła dla bezpośrednich połączeń
czas_przejazdu(X, Y, T) :-
    droga(X, Y, T).
% Pytanie i odpowiedź: ?- czas_przejazdu(warszawa, poznan, T). - T = 7

% Reguła obliczająca czas przejazdu przez połączenia pośrednie
czas_przejazdu(X, Y, T) :-
    droga(X, Z, T1),  % Sprawdzamy, czy istnieje droga z X do Z
    Z \== Y,  % Z nie może być Y, więc kontynuujemy przejazd
    czas_przejazdu(Z, Y, T2),  % Rekursja: obliczamy czas przejazdu z Z do Y
    T is T1 + T2.
```
